# Project Costs & Benefits Breakdown

You have spoken to various stakeholders in the company, including the operations team, the IT team and your manager. You have gathered the following information:

| Cost Category | Type | Amount (USD/year) | Notes |
|--------------|------|-------------------|--------|
| IT Infrastructure | Recurring | $21,000 | Cloud services, APIs, and system integration |
| Development | One-time | $45,000 | Direct labor costs for first year only |
| Maintenance | Recurring | $10,000 | Project maintenance costs |
| Current Operations | Baseline | $418,500 | Annual turbine maintenance |
| Current Losses | Baseline | $182,000 | Annual costs from unplanned downtime |


Additionally, your team has estimated that this project will result in:
-  a **20% reduction in downtime** and;
-  a **15% decrease in maintenance costs**.

Your task is to calculate the ROI for the first year of the project, following the steps and instructions provided below.


### Step 1: Calculate the total investment for the project for the first year

To calculate the total investment for the project, you need to add all the project costs for the first year. This includes the costs for cloud services, APIs and integration with existing systems, development costs, and maintenance costs.


```python
# Costs in the first year
it_infrastructure_costs = 21000
development_costs = 45000
maintenance_costs = 10000

# Total investment for the first year   
total_investment = it_infrastructure_costs + development_costs + maintenance_costs
```

### Step 2: Calculate the total benefits for the project for the first year

To calculate the total benefits for the project, you need to add all the project benefits for the first year. This includes the benefits from the project, such as the cost savings from unplanned downtime and maintenance costs.

Note: remember that you and your team have estimated that this project will result in a 20% reduction in downtime and a 15% decrease in maintenance costs. So you will need to take these into account when calculating the benefits.


```python

# Current operations costs  
current_annual_maintenance_costs = 418500
current_annual_losses_due_to_unplanned_downtime = 182000

# Yearly benefits
maintenance_cost_savings = current_annual_maintenance_costs * 0.15
unplanned_downtime_savings = current_annual_losses_due_to_unplanned_downtime * 0.20

# Total benefits for the first year
total_benefits = maintenance_cost_savings + unplanned_downtime_savings
```


### Step 3: Calculate the ROI for the first year

To calculate the ROI for the first year, you need to use the formula:

$$
ROI(\%) = \frac{Benefits - Total\;Investment}{Total\;Investment} \times 100
$$

Where:
- Benefits: The total gains from the project
- Total Investment: The total costs invested in the project




```python
# ROI Calculation
roi = (total_benefits - total_investment) / total_investment * 100

print(f"The ROI for the first year is {roi:.1f}%")
print(f"The total investment for the first year is ${total_investment:,.0f}")
print(f"The projected net benefits for the first year are ${total_benefits - total_investment:,.0f}")
print(f"The total benefits for the first year are ${total_benefits:,.0f}")
```

### 4. Stretch challenge: 

1. Calculate the ROI for the first 3 years of the project. Note that the development costs are one-time costs and the maintenance and IT costs are recurring costs.


### ROI Calculation: One-time Costs vs Yearly Costs

When calculating ROI over multiple years, you need to consider:
1. One-time costs (initial investments)
2. Recurring yearly costs
3. Expected yearly benefits

You can use the following formula:

$$
ROI(\%) = \frac{\sum_{t=1}^{n} Annual\;Benefits_t - (One\;Time\;Costs + \sum_{t=1}^{n} Yearly\;Costs_t)}{One\;Time\;Costs + \sum_{t=1}^{n} Yearly\;Costs_t} \times 100
$$

Where:
- t = year number
- n = total number of years

```python

# Calculate the ROI for the first 3 years

#One-time costs
one_time_costs = development_costs

#Recurring costs
recurring_costs = it_infrastructure_costs + maintenance_costs

#Expected yearly benefits
yearly_benefits = total_benefits

#Years
years = 3   

#Calculate the ROI for the first 3 years
roi = (yearly_benefits * years - (one_time_costs + (recurring_costs * years))) / (one_time_costs + (recurring_costs * years)   ) * 100

print(f"The ROI for the first 3 years is {roi:.1f}%")
print(f"The total investment for the first 3 years is ${one_time_costs + (recurring_costs * years):,.0f}")
print(f"The total benefits for the first 3 years are ${yearly_benefits * years:,.0f}")
print(f"The projected net benefits for the first 3 years are ${yearly_benefits * years - (one_time_costs + (recurring_costs * years)):,.0f}")
```
